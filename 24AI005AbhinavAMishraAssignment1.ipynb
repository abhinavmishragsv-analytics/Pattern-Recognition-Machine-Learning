Q15. Use Case: Linear Algebra
Solve the system of linear equations:
3x+2y+z=10
2x+3y+3z=18

Q13.Merging and Joining
Create two DataFrames:
o One with employee details (EmployeeID, Name, Department)
o One with salary details (EmployeeID, Salary)
Merge them using different types of joins: inner, outer, left, and right.

```python
employees = pd.DataFrame({
    'EmployeeID': [1, 2, 3],
    'Name': ['Amit', 'Ravi', 'Neha'],
    'Department': ['HR', 'IT', 'Finance']
})
salaries = pd.DataFrame({
    'EmployeeID': [1, 2, 4],
    'Salary': [50000, 60000, 70000]
})
inner_join = pd.merge(employees, salaries, on='EmployeeID', how='inner')
print(inner_join)
outer_join = pd.merge(employees, salaries, on='EmployeeID', how='outer')
print(outer_join)
left_join = pd.merge(employees, salaries, on='EmployeeID', how='left')
print(left_join)
right_join = pd.merge(employees, salaries, on='EmployeeID', how='right')
print(right_join)
```

Q4. Subplots
Create a figure with two subplots:
o Subplot 1: A sine wave plot for values of xxx from 000 to 22\pi2
o Subplot 2: A cosine wave plot for values of xxx from 000 to 22\pi2
Ensure both subplots have appropriate titles and axis labels.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, y_sin)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.subplot(1, 2, 2)
plt.plot(x, y_cos)
plt.title("Cosine Wave")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.tight_layout()
plt.show()
```
